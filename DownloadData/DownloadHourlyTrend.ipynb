```python
import os
import time

import datetime as datetime
from dateutil import relativedelta

import pandas as pd
from pytrends.request import TrendReq
```

```python
# Setting path of files
path_file_rawdata = r'C:\Users\patri\Google trends for buildings\Dataset\Hourly data\raw data'
path_file = r'C:\Users\patri\Google trends for buildings\Dataset\Hourly data'
```

```python
# Set list for keywords and geo_id
kw_list = ['/m/052zb', '/m/0dm258', '/m/0vlf']
geo_id_list = ['US', 'GB', 'IE', 'CA', 'CH', 'SG', 'TW']

# Download hourly data
year_start = 2015
year_end = 2020

for year in range(year_start, year_end+1):
    for month in range(1, 12+1):
        for geo_id in geo_id_list:
            try: 
                end_date = datetime.datetime(year,month,1) + relativedelta.relativedelta(months=1)
                name_file = str(year) +'-'+str(month) + '_' + str(end_date.year) +'-'+str(end_date.month) + '_' + geo_id + '.csv'
                if (not os.path.exists(os.path.join(path_file_rawdata,name_file))):                
                    print(name_file)

                    pytrends = TrendReq(hl='en-US', tz=360)

                    search_df = pytrends.get_historical_interest(kw_list, 
                                                                 year_start=year, month_start=month,
                                                                 year_end=end_date.year, month_end=end_date.month,
                                                                 geo=geo_id, gprop='', sleep=60)
                    search_df['geo_id'] = geo_id

                    if (len(search_df)>=24*28) and (pd.to_datetime(search_df.index[-1]).day==1) and (pd.to_datetime(search_df.index[0]).day==1):
                        search_df.to_csv(name_file)
                        print('Success!')
                        print('------------------------------------------')
                    else:
                        print('Incompleted dataframe : (')
                        print('------------------------------------------')
            except:
                print('error')               
                print('------------------------------------------')
                
```

```python
# Get the list of paths of hourly data (.csv)
list_csv = []

for file in os.listdir(path_file_rawdata):
    if file.endswith(".csv"):
        list_csv.append(os.path.join(path_file_rawdata, file))
        
list_csv
```

```python
# Merge and export the data
df_merged=pd.DataFrame()

for path_csv in list_csv:
    df_temp = pd.read_csv(path_csv)
    df_merged = pd.concat([df_merged,df_temp], ignore_index=True)
    
df_merged = df_merged.rename(columns={'/m/052zb':'Microsoft Office', '/m/0dm258':'Google Docs', '/m/0vlf':'Adobe'})
df_merged['date'] = pd.to_datetime(df_merged['date'])

df_merged.to_pickle(os.path.join(path_file, 'df_hourly_merged.pickle.gz'), compression='gzip')
df_merged
```
