```python
import os

import pytrends
from pytrends.request import TrendReq
from pytrends import dailydata

import numpy as np
import pandas as pd
```

```python
# Setting path of files
path_dataset = r'C:\Users\patri\Google trends for buildings\Dataset\Daily data'
```

```python
# Get keywords suggestions
pytrends = TrendReq(hl='en-US', tz=360)
suggs = pytrends.suggestions("microsoft office")
print(pd.DataFrame(suggs))
print('\n')

suggs = pytrends.suggestions("Google docs")
print(pd.DataFrame(suggs))
print('\n')

suggs = pytrends.suggestions("adobe")
print(pd.DataFrame(suggs))
print('\n')
```

```python
# Set list for keywords and geo_id
kw_list = ['Microsoft Office', 'Google Docs', 'Adobe']
kw_encode_list = ['/m/052zb', '/m/0dm258', '/m/0vlf']

geo_list = ['US', 'GB', 'IE', 'CA', 'CH', 'SG', 'TW']
#geo_list = ['US', 'US-FL', 'US-AZ', 'US-PA', 'US-TX', 'US-NV', 'US-MN', 'GB', 'GB-ENG', 'IE', 'CA', 'CA-QC']
```

```python
# Download daily data
list_df_daily = []
for keyword, keyword_encode in zip(kw_list, kw_encode_list):
    for geo_id in geo_list:
        df_daily = dailydata.get_daily_data(keyword_encode, 2016, 1, 2020, 7, geo = geo_id)
        df_daily['keyword'] = keyword
        df_daily['keyword_encode'] = keyword_encode
        df_daily['geo_id'] = geo_id
        df_daily.columns = df_daily.columns.str.replace(keyword_encode, 'trend')

        list_df_daily.append(df_daily)
```

```python
# Merge and export the data
df_daily_merged = pd.concat(list_df_daily, axis=0)
df_daily_merged.to_pickle(os.path.join(path_dataset, 'df_daily_merged.pickle'))
df_daily_merged
```
